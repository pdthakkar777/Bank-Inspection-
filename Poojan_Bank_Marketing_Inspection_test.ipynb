# Bank Marketing Data Analysis
This notebook presents an analysis of the `bankmarketing.csv` dataset, which includes data related to a bank's marketing campaigns. The main goal is to understand customer behavior and predict whether a client will subscribe to a term deposit.

```python

import pandas as pd

# Load the dataset
df = pd.read_csv('bankmarketing.csv')

# Display the first few rows
df.head()
```

```python

# Check for missing values and data types

df.isnull().sum()
```

```python
#Checking for data types and information of dataset 
df.info()
```

```python

# Summary statistics

df.describe().transpose()
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Plot the distribution of the target variable
```

```python
#Checking the counts of target variable
df['y'].value_counts()
```

```python
plt.figure(figsize=(8,7))
sns.countplot(df,
              x= 'y',
              color='darkblue',
              edgecolor='linen')
plt.title('The frequency distribution of positive and negative responses from campaign')
plt.xlabel('Yes/No')
plt.ylabel('Frequency')
plt.show()
```

#### Key Observation
#### “The campaign resulted in a 12% subscription rate, with 88% of customers choosing not to opt in.”

```python

# Correlation matrix for numerical features
correlation_matrix = df.select_dtypes('int64','float64').corr()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix,annot=True,cmap='viridis',fmt='.2f')
plt.title('Correlation amongst the numerical features')
plt.show()
```
